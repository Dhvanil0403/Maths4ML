```python
# Create a numpy array with the elements "1, 2, 3, 5, 0", print the array and test if all of its elements are non-zero.
array_A = np.array([1, 2, 3, 5, 0])
print(array_A)

non_zero_counter=0

for i in range(len(array_A)):
    if array_A[i]!=0:
        non_zero_counter+=1
        
    else:
        continue

print('There are ',non_zero_counter,'non zero values in the array')
```

```python
# Create a numpy array with the elements "3, -1, 3, 7, 2", print the array and test if any of its elements is non-zero.
array_B = np.array([3, -1, 3, 7, 2])
print(array_B)

non_zero_counter=0

for i in range(len(array_B)):
    if array_B[i]!=0:
        non_zero_counter+=1
        
    else:
        continue

print('There are ',non_zero_counter,'non zero values in the array')
```

```python
# Create a numpy array consisting of 100 random elements from normal distribution with mean of 2 and std of 0.5. 
#Print the mean, max, min and std of the array.

a = np.random.normal(loc=2, scale=0.5, size=100)

print(a)

a_mean = np.mean (a)
print ("the mean is:", a_mean)

a_max = np.max (a)
print ("the max is:", a_max)

a_min = np.min (a)
print ("the min is:", a_min)

a_std = np.std (a)
print ("the std is:", a_std)

```
